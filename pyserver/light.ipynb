```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
casual_df=pd.read_csv('Casualties0514.csv')
accident_df=pd.read_csv('Accidents0514.csv')
vehicle_df=pd.read_csv('Vehicles0514.csv')
```

```python
first_df=pd.merge(casual_df,accident_df,on='Accident_Index')
```

```python
df = first_df
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().values.any()
df.columns
```

```python
q3_df=pd.DataFrame(data=df,columns=['Longitude','Latitude','1st_Road_Class','2nd_Road_Class','Light_Conditions','Urban_or_Rural_Area','Weather_Conditions'])
```

```python
X=q3_df.drop('Light_Conditions',axis=1)
y=q3_df['Light_Conditions']
q3_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree= DecisionTreeClassifier()
```

```python
dtree.fit(X_train.values,y_train.values)
```

```python
predictions= dtree.predict(X_test.values)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import pickle
```

```python
print(classification_report(y_test.values,predictions))
```

```python
pickle.dump(dtree, open("light.pkl", "wb"))
```
