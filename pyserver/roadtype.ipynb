```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import pickle
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
accident_df=pd.read_csv('Accidents0514.csv')
```

```python
q3_df=pd.DataFrame(data=accident_df,columns=['Longitude','Latitude','Junction_Detail','Location_Easting_OSGR','Location_Northing_OSGR','Urban_or_Rural_Area','Road_Type'])
```

```python
q3_df
```

```python
X=q3_df.drop('Road_Type',axis=1)
y=q3_df['Road_Type']
q3_df.tail()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model= DecisionTreeClassifier()
```

```python
model.fit(X_train.values,y_train.values)
```

```python
predictions= model.predict(X_test.values)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
import numpy as np

# Define the features for a single observation
features = np.array([[-3.230826,54.985668,0,321337.0,566365.0,2]])

# Make a prediction
prediction = model.predict(features)

# Print the prediction
print("Prediction:", prediction)
```

```python
pickle.dump(model, open("road_type.pkl", "wb"))
```

```python
print(classification_report(y_test,predictions))
```
